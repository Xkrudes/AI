```python
"""Задание 1.3.1 Р/т №3"""
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

x = np.array([1, 2, 3])
y = np.array([4, 2, 3])
z = np.array([5, 2, 1])
w = np.array([6, 2, 4])

fig = plt.figure()
ax = fig.add_subplot(111, projection= '3d')

print(f'Расстояние Евклида между x и y = {np.linalg.norm(x-y)}')
print(f'Квадрат Евклидова расстояние между y и z = {np.linalg.norm(y-z) ** 2}')
print(f'Расстояние Чебышева между точками z и w = {np.linalg.norm(z-w, ord=np.inf)}')
print(f'Расстояние Хемминга между точками x и w = {np.linalg.norm(x-w, ord=1)}')

ax.scatter(1, 2, 3)
ax.scatter(4, 2, 3)
ax.scatter(5, 2, 1)
ax.scatter(6, 2, 4)
plt.show()
```

```python
"""Задание 2.3.1 Р/т №3"""
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

iris = sns.load_dataset('iris')

X_train, X_test, y_train, y_test = train_test_split(
    iris.iloc[:, :-1],
    iris.iloc[:, -1],
    test_size=0.15
)

model = KNeighborsClassifier(n_neighbors=10)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=iris,
    x='petal_width',
    y='petal_length',
    hue='species',
    s=70
)

plt.xlabel('Длина лепестка, см')
plt.ylabel('Ширина лепестка, см')
plt.legend(loc=2)
plt.grid()

for i in range(len(y_test)):
    if np.array(y_test)[i] != y_pred[i]:
        plt.scatter(X_test.iloc[i, 3], X_test.iloc[i, 2], color='red', s=150)

print(f'accuracy: {accuracy_score(y_test, y_pred) :.3}')
```
