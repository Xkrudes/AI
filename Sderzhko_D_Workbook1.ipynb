```python
"""Задание 4.3.2 Р/т №1"""
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 10, 10)

def my_function(x):
    return (1/x + 1/x**2 - 2/x**3 + 2/x**4 + 3*np.cos(x) - np.sin(x) - np.log(x))

y = my_function(x)

x_slice = x[:5]
y_slice = y[:5]


plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Линейный график для основного массива')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.subplot(122)
plt.plot(x_slice, y_slice, marker='o', linestyle='', color='r')
plt.title('Точечный график для среза')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.tight_layout()
plt.show()
```

```python
"""Задание 4.3.3 Р/т №1"""
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapz


x = np.arange(0, 11, 1)

def my_function(x):
    return np.abs(np.cos(x * (np.cos(x) + np.log(x + 1))))

y = my_function(x)

plt.plot(x, y, label='f(x)')
plt.fill_between(x, y, alpha=0.3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции с заливкой площади')
plt.legend()

area = trapz(y, x)
print(f'Площадь под графиком: {area:.2f}')

plt.show()
```

```python
"""Задание 4.3.5 Р/т №1"""
import numpy as np

def add(x, y):
    return x + y
def subtract(x, y):
    return x - y
def multiply(x, y):
    return x * y
def divide(x, y):
    if y == 0:
        return "Ошибка: деление на ноль"
    return x / y
def exponential(x, y):
    return np.exp(x + y)
def sine_cosine(x, y):
    return np.sin(x + y), np.cos(x + y)
def power(x, y):
    return x ** y

for i in range(1):
    print("Выберите операцию:")
    print("1. Сложение")
    print("2. Вычитание")
    print("3. Умножение")
    print("4. Деление")
    print("5. Вычисление e^(x + y)")
    print("6. Вычисление sin(x + y) и cos(x + y)")
    print("7. Вычисление x^y")
    print("8. Выход")

    choice = input("Введите номер операции (1/2/3/4/5/6/7/8): ")

    if choice == '8':
        print("До свидания!")
        break

    if choice in ('1', '2', '3', '4', '5', '6', '7'):
        x = float(input("Введите первое число: "))
        y = float(input("Введите второе число: "))

        if choice == '1':
            print("Результат:", add(x, y))
        elif choice == '2':
            print("Результат:", subtract(x, y))
        elif choice == '3':
            print("Результат:", multiply(x, y))
        elif choice == '4':
            print("Результат:", divide(x, y))
        elif choice == '5':
            print("Результат:", exponential(x, y))
        elif choice == '6':
            sin_result, cos_result = sine_cosine(x, y)
            print("sin(x + y) =", sin_result)
            print("cos(x + y) =", cos_result)
        elif choice == '7':
            print("Результат:", power(x, y))
    else:
        print("Неправильный ввод. Попробуйте ещё раз.")
```
